```python
import torch, torchvision
from torch import nn
from torch import optim
from torchvision.transforms import ToTensor
import torch.nn.functional as F
import matplotlib.pyplot as plt
import random as r
import copy
import numpy as np
```

Zbiór testowy ma 10000 próbek, uczący 60000. Dla batcha wymiaru 16 mamy: 625 i 3750 zbiorów.

```python
numb_batch = 1 #to ensure equal chances with Julia number of batches has to be 1
```

Import bazy testującej i uczącej oraz ich załadowanie w postaci tensora. Z obrazka 28x28 do tensora 784x1.

```python
T = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor()
])
train_data = torchvision.datasets.MNIST('mnist_data', train=True, download=True, transform=T)
test_data = torchvision.datasets.MNIST('mnist_data', train=False, download=True, transform=T)

train_dl = torch.utils.data.DataLoader(train_data, batch_size = numb_batch)
test_dl = torch.utils.data.DataLoader(test_data, batch_size = numb_batch)
```

```python
def create_net():
    model = nn.Sequential(
        #lenet for performance testing
        #nn.Conv2d(1, 6, 5, padding=2),
        #nn.ReLU(),
        #nn.AvgPool2d(2, stride=2),
        
        #nn.Conv2d(6, 16, 5, padding=0),
        #nn.ReLU(),
        #nn.AvgPool2d(2, stride=2),
        
        #nn.Flatten(), #warstwa spłaszczająca do wektora
        #nn.Linear(400, 120),
        #nn.ReLU(),
        #nn.Linear(120, 84),
        #nn.ReLU(),
        #nn.Linear(84, 10),
        
        # Same exact model as it was implemented in Julia
        nn.Conv2d(1, 1, 5, padding=2), #1, 6, 5, padding=0, czyli output size był 24x24. Padding 2 daje output 28x28 czyli bez zmian wymiarów
        #nn.Sigmoid(),
        nn.AvgPool2d(2, stride=2),
        nn.Flatten(), #warstwa spłaszczająca do wektora
        nn.Linear(196, 120), #864
        nn.Sigmoid(),
        nn.Linear(120, 84),
        nn.Sigmoid(),
        nn.Linear(84, 10),
    )
    return model
```

Funkcja testująca model na zadanym zbiorze danych. Używana podczas trenowania sieci, aby zdobyć informacje o jej skuteczności.


```python
def test(model, data):
    total = 0
    correct = 0
    for i, (images, labels) in enumerate(data):
        #images = images.cuda()
        x = model(images)
        value, pred = torch.max(x,1)
        #pred = pred.data.cpu()
        total += x.size(0)
        correct += torch.sum(pred == labels)
    return correct*100./total
```

Poniżej znajduje się funkcja trenująca sieć. Oprócz inicjalizacji modelu zaimplementowana logika zapewnia zwrócenie specyfikacji najlepszego modelu wygenerowanego podczas trenowania epok. Inne metody z biblioteki wyjaśniono poniżej.

nn.CrossEntropyLoss() -- metoda definiująca funkcję strat zwaną entropią krzyżową. Wykorzystuje wzór przypisujący prawdopodobieństwo przynależności do klasy na podstawie funkcji logarytmicznej. Jest korzystna przy wstecznej propagacji błędu, ponieważ gdy prawdopodobieństwo jest bliskie zeru to styczna w tym punkcie (pochodna) jest mocno nachylona co pozwala na skuteczną zmianę wartości wag.

optim.Adam() -- metoda optymalizacji (adaptive moment estimation) łącząca cechy RMSprop i Momentum. Zazwyczaj osiągająca od nich lepsze wyniki, przez co jest często używana. Domyślne bety w bibliotece Pytorch to beta1=0.9 i beta2=0.999.

optimizer.zero_grad() -- metoda służąca do wyzerowania gradientów optimizera (w moim przypadku Adama). Należy to wykonać przy każdej nowej iteracji w zestawie trenującym. Wymagana, aby potem użyć metody backward().

loss.backward() -- oblicza gradient funkcji strat z pomocą wstecznego automatycznego różniczkowania. Przydatne linki: http://karpathy.github.io/neuralnets/ oraz https://discuss.pytorch.org/t/what-does-tensor-backward-do-mathematically/27953/4 oraz https://pytorch.org/docs/stable/_modules/torch/autograd.html#backward

optimizer.step() -- metoda wykonująca pojedynczy krok optymalizacji wag. Należy umieszczać ją po funkcji liczenia gradientu funkcji strat.

```python
from tqdm import tqdm
def train(numb_epoch=2, lr=1e-3, device="cpu"):
    accuracies = []
    cnn = create_net().to(device)
    cec = nn.CrossEntropyLoss()
    optimizer = optim.Adam(cnn.parameters(), lr=lr)
    max_accuracy = 0
    for epoch in range(numb_epoch):
        for i, (images, labels) in tqdm(enumerate(train_dl), desc=f"EPOCH {epoch+1}"):
            images = images.to(device)
            labels = labels.to(device)
            optimizer.zero_grad()
            pred = cnn(images)
            loss = cec(pred, labels)
            loss.backward()
            optimizer.step()
        accuracy = float(test(cnn, test_dl))
        accuracies.append(accuracy)
        if accuracy > max_accuracy:
            best_model = copy.deepcopy(cnn)
            max_accuracy = accuracy
            print("Saving Best Model with Accuracy: ", accuracy, '%')
        print('Epoch:', epoch+1, "Accuracy :", accuracy, '%')
    plt.plot(accuracies)
    return best_model
```

```python
if torch.cuda.is_available():
    device = torch.device("cuda:0")
else:
    device = torch.device("cpu")
```

Block responsible for memory tracking.

```python
import tracemalloc

tracemalloc.start()
net = train(numb_epoch=1, lr=1e-2)
print(tracemalloc.get_traced_memory())
tracemalloc.stop()
```

```python
#if memory tracing block is not run, copile this to see accuracies
net = train(numb_epoch=3, lr=1e-2)
```

Funkcja do predykcji modelu na zadanym zbiorze. Zwraca listę z predykcjami i listę z etykietami ze zbioru testowego.

```python
def predict(model, data):
    y_pred = []
    y_true = []
    for i, (images, labels) in enumerate(data):
        #images = images.cuda()
        x = model(images)
        value, pred = torch.max(x, 1)
        #pred = pred.data.cpu()
        y_pred.extend(list(pred.numpy()))
        y_true.extend(list(labels.numpy()))
    return np.array(y_pred), np.array(y_true)
```

```python
y_predicted, y_true = predict(net, test_dl)
```
